## numpy 행렬 곱 (dot product) (I)
- 행렬 곱(dot product)<br>
=> A 행렬과 B 행렬의 행렬 곱(dot product)는 np.dot(A,B)로 나타내며, 행렬 A의 열 벡터와 B 행렬의 행 벡터가 같아야 함. 만약 같지 않다면 reshape 또는 전치행렬(transpose) 등을 사용하여 형 변환을 한 후에 행렬 곱을 실행해야 함

```python
import numpy as np
```

```python
A = np.array([ [1, 2, 3], [4, 5, 6] ]) # 2X3 행렬
B = np.array([ [-1, -2], [-3, -4], [-5, -6] ]) # 3X2 행렬

# (2X3) dot product (3X2) == (2X2) 행렬
C = np.dot(A, B) # 행렬 곱 수행

# matrix A, B 형상 출력 => shape
print("A.shape ==", A.shape, ", B.shape ==", B.shape)
print("C.shape ==", C.shape)
print(C)
```

## numpy 행렬 곱 (dot product) (II)
- 행렬 곱(dot product)<br>
    => 행렬 곱은, 행렬의 원소 개수가 같아야만 계산할 수 있는 사칙연산의 한계를 벗어나<br>
    1. 행렬곱 조건을 만족하는 다양한 크기의 행렬을 연속으로 만들고
    2. 행렬 곱을 연속으로 계산하면서
    3. 결과값을 만들 수 있기 때문에 머신러닝과 이미지 프로세싱 분야에서 자주 사용됨
    
    ex) 입력 행렬 형상이 64X64이고, 결과 행렬 형상이 64X10이라면, 중간에 dot product를 만족시키는 어떤 형상의 행렬이라도 가질 수 있음.
    
행렬 곱을 사용하지 않고 산술 연산만 가능하다면, 입력 행렬의 64X64 크기를 가지는 특성 값만을 사용해야 하기 때문에 다양한 특성을 갖는 필터 개발이 불가능함.

(64 X 64) * (64 X 256) * (256 X 512) * (512 X 64) * (64 X 10) = (64 X 10)

## numpy broadcast
- 행렬의 사칙연산은 기복적으로 두 개의 행렬 크기가 같은 경우에만 수행할 수 있음. 그러나 numpy에서는 크기가 다른 두 행렬 간에도 사칙연산(+, -, *, /)을 할 수 있는데, 이를 브로드캐스트(broadcast)라고 지칭.<br>

    => 차원이 작은 쪽이 큰 쪽의 행 단위로 반복적으로 크기를 맞춘 후에 계산
    
    *broadcast는 행렬곱 연산이 아닌, '사칙연산'에만 적용*

```python
A = np.array([ [1, 2], [3, 4] ])
b = 5 # b는 scala

print(A+b)
```

```python
C = np.array([ [1, 2], [3, 4] ])
D = np.array([4, 5])

print(C+D)
```

## numpy 전치행렬 (transpose)
- 전치행렬 (transpose)<br>

=> 어떤 행렬의 전치행렬 (transposed matrix)은 원본 행렬의 열은 행으로, 행은 열로 바꾼 것이로서, 원본 행렬을 A라고 하면 전치행렬은 A^T로 나타냄.

즉, 1행은 1열로 바꾸고 2행은 2열로, 3행은 3열로 바꾼 행렬을 의미

```python
A = np.array([ [1, 2], [3, 4], [5, 6] ]) # 3X2 행렬

B = A.T # A의 전치행렬, 2X3 행렬

print("A.shape ==", A.shape, ", B.shape ==", B.shape)
print()
print(A)
print()
print(B)
```

```python
# vector 전치행렬

C = np.array([1, 2, 3, 4, 5]) # vector, matrix 아님
D = C.T # C는 vector이므로 transpose 불가

E = C.reshape(1, 5)
F = E.T # E의 전치행렬

print("C.shape ==", C.shape, ", D.shape ==", D.shape)
print("E.shape ==", E.shape, ", F.shape ==", F.shape)
print(F)
```

## numpy 행렬 indexing / slicing
- 행렬 원소 접근 (I)<br>
=> 행렬 원소를 명시적(explicit)으로 접근하기 위해서는 *리스트(list)*에서처럼, 인덱싱 / 슬라이싱 모두 사용 가능함.

[예제 1] A[0, 0] 또는 A[0][0]은 1행 1열, A[2, 1], 또는 A[2][1]은 3행 2열임

```python
A = np.array([10, 20, 30, 40, 50, 60]).reshape(3, 2)

print("A.shape ==", A.shape)
print(A)
```

[예제 2] A[:, 0]은 모든 행 1열을 나타냄

```python
print("A[0, 0] ==", A[0, 0], ", A[0][0] ==", A[0][0])
print("A[2, 1] ==", A[2, 1], ", A[2][1] ==", A[2][1])
```

[예제 3] A[0:-1, 1:2] 인덱스 0인 1행부터, 인덱스 -1-1=-2인 2행까지의 모든 데이터, 그리고 인덱스 1인 2열부터 인덱스 2-1=1인 2열까지의 모든 데이터

```python
print("A[0:-1, 1:2] ==", A[0:-1, 1:2])
```

[예제 4] A[:, :]은 모든 행, 모든 열

```python
print("A[ :, 0] ==", A[:, 0])
print("A[ :, :] ==", A[:, :])
```
