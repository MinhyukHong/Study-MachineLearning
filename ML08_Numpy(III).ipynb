## numpy concatenate
- 행렬에 행(row) 또는 열(column)을 추가하기 위한 numpy.concatenate(...)<br>
=> 머신러닝의 회귀(regression)코드 구현 시 가중치(weight)와 바이어스(bias)를 별도로 구분하지 않고, 하나의 행렬로 취급하기 위한 프로그래밍 구현 기술

```python
import numpy as np
```

```python
# 행렬에 열과 행 추가

A = np.array([ [10, 20, 30], [40, 50, 60] ])

print(A.shape)

# A matrix에 행(row)을 추가할 행렬. 1행 3열로 reshape
# 행을 추가하기 때문에 우선 열을 3열로 만들어야 함
# 벡터는 행과 열을 구분하지 않는 1차원 배열이기 때문에, reshape을 해주지 않으면 행렬의 특성을 가질 수 없음
row_add = np.array([70, 80, 90]).reshape(1, 3)
print(row_add.shape)

# A matrix에 열(column)을 추가할 행렬. 2행 1열로 reshape
# 열을 추가하기 때문에 우선 행을 2행으로 만들어야 함
column_add = np.array([1000, 2000]).reshape(2, 1)
print(column_add.shape)

# numpy.concatenate에서 axis=0 행(row) 기준
# A 행렬에 row_add 행렬 추가
B = np.concatenate((A, row_add), axis=0)

print(B)

# numpy.concatenate에서 axis=1 열(column) 기준
# B 행렬에 column_add 행렬 추가
C = np.concatenate((A, column_add), axis=1)

print(C)
```

## numpy useful function (I)
- seperator로 구분된 파일에서 데이터를 읽기 위한 numpy.loadtxt(...)

[예제] 다음과 같이 콤마(,)로 분리된 데이터 파일을 read하기 위해서는 np.loadtxt("파일명", seperator=',')를 호출함.<br>
*리턴값은 행렬이기 때문에 인덱싱 또는 슬라이싱을 이용하여 데이터를 분리할 필요 있음.*<br>
=> 머신러닝 코드에서 입력데이터와 정답데이터를 분리하는 프로그래밍 기법

```python
loaded_data = np.loadtxt('/Users/LG/Desktop/z.MachineLearing_Study/data-01.csv', delimiter=',', dtype=np.float32)

x_data = loaded_data[ :, 0:-1]
t_data = loaded_data[ :, [-1]]

# 데이터 차원 및 shape 확인
print("x_data.ndim = ", x_data.ndim, ", x_data.shape = ", x_data.shape)
print("t_data.ndim = ", t_data.ndim, ", t_data.shape = ", t_data.shape)
```

## numpy useful function (II)

- numpy.random.rand(...)

```python
# 0 ~ 1 사이의 random number 발생

random_number1 = np.random.rand(3)
random_number2 = np.random.rand(1, 3)
random_number3 = np.random.rand(3, 1)

print("random_number1 ==", random_number1, ", random_number1.shape ==", random_number1.shape)
print("random_number2 ==", random_number2, ", random_number2.shape ==", random_number2.shape)
print("random_number3 ==", random_number3, ", random_number3.shape ==", random_number3.shape)
```

- numpy.sum(...), numpy.exp(...), numpy.log(...)

```python
X = np.array([2, 4, 6, 8])

print("np.sum(X) ==", np.sum(X))
print("np.exp(X) ==", np.exp(X))
print("np.log(X) ==", np.log(X))
```

## numpy useful function (III)

- numpy.max(...), numpy.min(...), numpy.argmax(...), numpy.argmin(...)<br>

*argmax(), argmin()은 각 최대값, 최솟값이 있는 곳의 index를 return*<br>
=> max(), argmax() 함수는 classification(분류기준)에서 정답의 위치를 찾아내는 데에 있어 중요한 개념

```python
# 벡터 적용 예

X = np.array([2, 4, 6, 8])

print("np.max(X) ==", np.max(X))
print("np.min(X) ==", np.min(X))
print("np.argmax(X) ==", np.argmax(X))
print("np.argmin(X) ==", np.argmin(X))
```

```python
# 행렬 적용 예
# 행(row) 혹은 열(column) 중 기준을 정해주어야 함

X = np.array([ [2, 4, 6], [1, 2, 3], [0, 5, 8] ])

print("np.max(X) ==", np.max(X, axis=0)) # axis=0, 열기준
print("np.min(X) ==", np.min(X, axis=0)) # axis=0, 열기준

print("np.max(X) ==", np.max(X, axis=1)) # axis=1, 행기준
print("np.min(X) ==", np.min(X, axis=1)) # axis=1, 행기준

print("np.argmax ==", np.argmax(X, axis=0)) # axis=0, 열기준
print("np.argmin ==", np.argmin(X, axis=0)) # axis=0, 열기준

print("np.argmax ==", np.argmax(X, axis=1)) # axis=1, 행기준
print("np.argmin ==", np.argmin(X, axis=1)) # axis=1, 행기준
```

- numpy.ones(...), numpy.zeros(...)<br>

*ones()는 주어진 형상의 행렬을 1로 채우고, zeros()는 주어진 형상의 행렬을 0으로 채움*

```python
A = np.ones([3, 3])

print("A.shape ==", A.shape, ", A ==", A)

B = np.zeros([3, 2])

print("B.shape ==", B.shape, ", B ==", B)
```

=> 입력 데이터의 분포와 모양을 먼저 그래프로 그려보고, 데이터의 특성과 분포를 파악한 후 어떤 알고리즘을 적용할 지 결정하고 있음<br>

## scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

# 주피터 노트북을 사용하는 경우 노트북 내부에 그림 표시
%matplotlib inline

# x data, y data 생성
x_data = np.random.rand(100)
y_data = np.random.rand(100)

plt.title('scatter plot')
plt.grid()
plt.scatter(x_data, y_data, color='b', marker='o')
plt.show()
```

## line plot

```python
import matplotlib.pyplot as plt

%matplotlib inline

# list comprehension을 통한 list 생성
x_data = [ x for x in range(-5,5) ]
y_data = [ y*y for y in range(-5,5) ]

plt.title('line plot')
plt.grid()
plt.plot(x_data, y_data, color='b')
plt.show()
```
